# Pytorch (Native Pipeline Flow)

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Making the dataframe using the csv file of Breast Cancer
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
# Presenting the Header of Dataframe
df.head()
# Header gives us good insights of the complete data means how many
# columns are there, which column is unnecessary or don't needed and
# which column needs some handling before use and so
# Like in this case we don't need 'id' and 'unnamed' columns so we will
# dropout them in near future
```

```python
# Checking the Shape of Dataframe
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
df.head()
```

```python
# Splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2)
```

```python
# Scaling the data using StandarScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Printing the Scaled data and labels
print(X_train)
print(y_train)
```

```python
# NOTE THE LABELS ARE OF 2 ALPHABET CLASSES (M AND B) SO FIRST
# ENCODING THEM USING LABELENCODER
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# Now Printing the Coded Labels
print(y_train)
```

```python
# NOW WE WANT TO MAKE OUR MODEL IN PYTORCH BUT WE HAVE EVERYTHING IN
# NUMPY ARRAYS SO CONVERTING THEM
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
# Now Printing the Shapes of the tensors
print(X_train_tensor.shape)
print(X_test_tensor.shape)
print(y_train_tensor.shape)
print(y_test_tensor.shape)
```

```python
import torch
import torch.nn as nn

class MySimpleNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
# Initializing the Important Parameters for our model
# Epochs - No of Iterations for which model is run (usually for trained)
epochs = 25
```

```python
# Model Creation
model = MySimpleNN(X_train_tensor.shape[1])
```

```python
# Loss Function and Optimizer Formation
loss_function = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(epochs):
  # Cast X_train_tensor to float32 to match model's expected input dtype
  # Cast y_train_tensor to float32 for the loss calculation
  y_pred = model(X_train_tensor.float())
  loss = loss_function(y_pred, y_train_tensor.view(-1,1).float())
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")
```

```python
# NOW EVALUATING THE MODEL BY TESTING IT WITH OUR TESTING DATA
# NOTE AT MODEL TESTING LEVEL WE ARE TESTING THAT THE MODEL IS TRAINED
#   OR NOT SO WE WILL NOT BE UPDATING WEIGHTS (NO BACKPROPAGATION)
#   INSTEAD WILL BE CHECKING OUR TRAINED WEIGHTS AND BIAS
#   SO WE WILL MAKE THE AUTOGRAD TRACKING OFF AND WILL USE
#   THE TRAINED MODEL WEIGHTS, BIAS AND EVERY OTHER THING AND
#   THEN WE WILL CALCULATE THE DIFFERENCE WITH ACTUAL VALUES
#   AND THIS WHOLE PROCESS IS CALLED EVALUATION.
with torch.no_grad():
  y_pred_test = model.forward(X_test_tensor.float())
  # In this logistic regression model, forward pass is giving values
  # as mapped probabilities ranging from 0 to 1 but our original labels
  # are encoded as 0 and 1 so we will be putting a thershold or decision
  # boundary for our class 0 and class 1 (manually here)
  # NOTE CHANGIN THE THERSHOLD WILL CHANGE THE ACCURACY ACCORDINGLY
  y_pred_test = (y_pred_test > 0.5).int()
  # Accuracy is obtained manually
  accuracy = (y_pred_test == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy}")
```
